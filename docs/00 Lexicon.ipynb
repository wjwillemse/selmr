```python
import pickle
with open('..//data//dbpedia_phrases_10000.pickle', 'rb') as handle:
    phrases = pickle.load(handle)
with open('..//data//dbpedia_contexts_10000.pickle', 'rb') as handle:
    contexts = pickle.load(handle)
```

```python
# Creating the NLP processor
import stanza

lang = "en"

# create a Stanza pipeline for pretokenized data
nlp = stanza.Pipeline(
        lang=lang, 
        processors='tokenize, lemma', 
        tokenize_pretokenized=True,
        download_method=None,
        verbose=False
)
```

```python
text = [item.split(" ") for item in list(phrases.keys())]
```

```python
stanza = nlp(text)
```

```python
from collections import defaultdict

lemmas = defaultdict(set)
for sentence in stanza.sentences:
    if all([word.lemma is not None for word in sentence.words]):
        text = " ".join([word.text for word in sentence.words])
        lemma = " ".join([word.lemma for word in sentence.words]).replace("_", "")
        lemmas[text].add(lemma)
```

```python
from collections import Counter

for lemma, ps in lemmas.items():
    vector = Counter()
    for p in ps:
        if p in phrases.keys():
            vector.update(phrases[p])
    for p in ps:
        if p in phrases.keys():
            phrases[p+"[lemma]"] = vector
```
