# Working with multisets

```python
import logging, sys
logging.basicConfig(stream=sys.stdout,
                    format='%(asctime)s %(message)s',
                    level=logging.DEBUG)
```

```python
import pickle
from selmr import LanguageMultisets
with open('..//data//dbpedia_phrases_10000.pickle', 'rb') as handle:
    phrases = pickle.load(handle)
with open('..//data//dbpedia_contexts_10000.pickle', 'rb') as handle:
    contexts = pickle.load(handle)
with open('..//data//dbpedia_lemmas_10000.pickle', 'rb') as handle:
    lemmas = pickle.load(handle)
multisets = LanguageMultisets(phrases, contexts, lemmas)
```

```python
from selmr import SELMR, LanguageMultisets

# construct a SELMR data structure with the DBpedia phrases and contexts
selmr = SELMR(
    multisets=multisets
)
```

```python
m1 = selmr.contexts('smaller', topn=15)
m2 = selmr.contexts('larger', topn=15)
```

```python
from selmr import jaccard_index

# calculate the Jaccard index between m1 and m2
jaccard_index(m1, m2)
```

So len(m1 & m2) equals 11 and len(m1 | m2) equals 19.

```python
# Create a multiset with the top 15 of the intersection of m1 and m2
(m1 & m2).topn(15)
```

```python
(m1 - m2).topn(15)
```

```python
m1 = selmr.contexts("river", include_other_forms=True, topn=None)
m2 = selmr.contexts("river", topn=None)
m3 = selmr.contexts("rivers", topn=None)
```

```python
from selmr import containment_index
```

```python
containment_index(s1, s3)
```
